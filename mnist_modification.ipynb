```python
import numpy as np
import tensorflow as tf

mnist = tf.keras.datasets.mnist
(x_train_orig, y_train), (x_test_orig, y_test) = mnist.load_data()

x_train = np.zeros((x_train_orig.shape[0], 50, 50), dtype=np.uint8)
x_test = np.zeros((x_test_orig.shape[0], 50, 50), dtype=np.uint8)

for i in range(len(x_train_orig)):
    mnist_digit = x_train_orig[i]
    background_image = np.random.randint(0, 256, (50, 50))
    background_image[12:40, 15:43] = mnist_digit
    x_train[i] = background_image

for i in range(len(x_test_orig)):
    mnist_digit = x_test_orig[i]
    background_image = np.random.randint(0, 256, (50, 50))
    background_image[12:40, 15:43] = mnist_digit
    x_test[i] = background_image
```

```python
from tensorflow.keras.layers import Flatten, Dense
from tensorflow.keras.models import Sequential

x_train_normalized = x_train / 255.0
x_test_normalized = x_test / 255.0

model = Sequential([
  Flatten(input_shape=(50, 50)),
  Dense(32, activation='relu'),
  Dense(32, activation='relu'),
  Dense(10, activation='softmax')
])

model.compile(optimizer='adam',
              loss='sparse_categorical_crossentropy',
              metrics=['accuracy'])

model.fit(x_train_normalized, y_train, epochs=1, validation_data=(x_test_normalized, y_test))
```

```python
E1=set()
for i in range(50):
  for j in range(50):
    flag=True
    if(12<=i<40) and (15<=j<43):
      flag=False
    if(flag):
      E1.add((i,j))
```

```python
def dist(p1,p2):
  return ((p1[0]-p2[0])**2+(p1[1]-p2[1])**2)**(0.5)

v1=(12,15)
v2=(39,15)
v3=(39,42)
v4=(12,42)

E2=[]
for i in range(50):
  for j in range(50):
    if (i,j) not in E1:
      x=(i,j)
      E2.append([x,min(dist(x,v1),dist(x,v2),dist(x,v3),dist(x,v4)),min(i-12,39-i,j-15,42-j)])

E2.sort(key=lambda y: [y[1],y[2]])
```
